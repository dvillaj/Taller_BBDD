# Data Types: Maps

Con los buckets de tipo Mapa, Riak permite guardar datos con estructura!

```python
import riak

client = riak.RiakClient()
```

## Create a map bucket

Para este ejemplo, vamoa a creat  un bucket de tipo mapa para guardar las propiedades de nuestros clientes.

Un mapa contiene registros de distintos tipos. En el ejemplo utilizaremos un registro de tipo contador

```python
bucket_customers = client.bucket_type('maps').bucket('customers')
```

## Insert Data

```python
customer_map = bucket_customers.new('ahmed_info')

customer_map.registers['first_name'].assign('Ahmed')
customer_map.registers['phone_number'].assign('5551234567')

# Counters
customer_map.counters['page_visits'].increment(1)
customer_map.store()
```

## Retrieve Data

```python
customer_map = bucket_customers.get('ahmed_info')

print(customer_map.registers['first_name'])
print(customer_map.registers['phone_number'])
print(customer_map.counters['page_visits'].value)
```
